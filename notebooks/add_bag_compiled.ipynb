```python
import pandas as pd 
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
file_path = '../data/compiled_results/raw/Compiled Field Results - CFTP Gathered Data.xlsx'
xls = pd.ExcelFile(file_path)
all_trials = {sheet_name: pd.read_excel(xls, sheet_name=sheet_name, skiprows=None if sheet_name != xls.sheet_names[0] else 1)
              for sheet_name in xls.sheet_names}
```

```python
facility_ad = all_trials['AD001-01']
facility_ad
```

```python
def get_bag_info(facility):
    df = all_trials[facility]
    columns_to_keep = ['Trial ID', 'Trial Bag Set', 'Trial Bag ID', 'Load Concentration']
    df = df[columns_to_keep]
    df.columns = [col.lower() for col in df.columns]
    df.columns = df.columns.str.replace(' ', '_')
    df = df.drop_duplicates(subset='trial_bag_id', keep='first')
    df['trial_bag_set'] = df['trial_bag_set'].str.replace('Set ', '')
    df['trial_bag_id'] = df['trial_id'] + '_' + df['trial_bag_id']
    new_names = {
        'trial_bag_set': 'bag_set',
        'trial_bag_id' : 'bag_id'
    }
    df = df.rename(columns=new_names)
    return df
```

```python
facility_ad = get_bag_info('AD001-01')
facility_wr01 = get_bag_info('WR001-01')
facility_casp01 = get_bag_info('CASP001-01')
```

```python
facility_wr01['bag_id'] = facility_wr01['bag_id'].str.split().str[0]
facility_ad['bag_id'] = facility_ad['bag_id'].str.replace(' ', '-')
facility_casp01['bag_id'] = facility_casp01['bag_id'].str.replace(' ', '-')
```

```python
facility_wr01
```

```python
facility_casp03 = all_trials['CASP003-01']
columns_to_keep = ['Trial ID', 'Trial Bag ID', 'Trial Bag Colour', 'Bag Set Detail']
facility_casp03 = facility_casp03[columns_to_keep]
facility_casp03.columns = [col.lower() for col in facility_casp03.columns]
facility_casp03.columns = facility_casp03.columns.str.replace(' ', '_')
facility_casp03 = facility_casp03.drop_duplicates(subset='trial_bag_id', keep='first')
facility_casp03['trial_bag_id'] = facility_casp03['trial_id'] + '_' + facility_casp03['trial_bag_id']
new_names = {
        'trial_bag_colour': 'bag_colour',
        'trial_bag_id' : 'bag_id'
    }
facility_casp03 = facility_casp03.rename(columns=new_names)
facility_casp03
```

```python
facility_wr03 = all_trials['WR003-01']
columns_to_keep = ['Trial ID', 'Trial Bag ID', 'Trial Bag Colour']
facility_wr03 = facility_wr03[columns_to_keep]
facility_wr03.columns = [col.lower() for col in facility_wr03.columns]
facility_wr03.columns = facility_wr03.columns.str.replace(' ', '_')
facility_wr03 = facility_wr03.drop_duplicates(subset='trial_bag_id', keep='first')
facility_wr03['trial_bag_id'] = facility_wr03['trial_bag_id'].str.replace('#', '').str.replace(' ', '-')
facility_wr03['trial_bag_id'] = facility_wr03['trial_id'] + '_' + facility_wr03['trial_bag_id']
new_names = {
        'trial_bag_colour': 'bag_colour',
        'trial_bag_id' : 'bag_id'
    }
facility_wr03 = facility_wr03.rename(columns=new_names)
facility_wr03
```

```python
merged_cftp = pd.concat([facility_ad, facility_wr01, facility_casp01, facility_casp03, facility_wr03], ignore_index=True)
merged_cftp
```

```python
compiled_10 = pd.read_csv('../data/compiled_results/processed/observation_mass.csv')
compiled_10 = compiled_10[compiled_10['bag_number'] != '(blank)']
split_columns = compiled_10['bag_set'].str.split(' ', n=1, expand=True)
compiled_10['bag_set'] = split_columns[0]
compiled_10['bag_colour'] = split_columns[1].str.replace(r"[()]", "", regex=True)
compiled_10['bag_colour'] = compiled_10['bag_colour'].str.capitalize()
compiled_10['bag_id'] = compiled_10['facility_id'] + '_' + compiled_10['bag_set'] + compiled_10['bag_number'].astype(str)
columns_to_keep = ['facility_id', 'bag_set', 'bag_id', 'bag_colour']
compiled_10 = compiled_10[columns_to_keep]
new_names = {
        'facility_id': 'trial_id',
    }
compiled_10 = compiled_10.rename(columns=new_names)
compiled_10
```

```python
merged_compiled = pd.concat([merged_cftp, compiled_10], ignore_index=True)
merged_compiled
```

```python
casp_004 = pd.read_csv('../data/CASP004-01/masterfile.csv')
columns_to_keep = ['trial_ID', 'bag_color', 'bag_set', 'bag_ID', 'bag_content', 'bag_type', 'bag_placement']
casp_004 = casp_004[columns_to_keep]
casp_004['trial_ID'] = 'CASP004-01'
casp_004['bag_ID'] = casp_004['trial_ID'] + '_' + casp_004['bag_ID']
new_names = {
        'bag_color': 'bag_colour',
    }
casp_004 = casp_004.rename(columns=new_names)
casp_004.columns = casp_004.columns.str.lower()
casp_004
```

```python
all_bags = pd.concat([merged_compiled, casp_004], ignore_index=True)
all_bags.reset_index(inplace=True)
all_bags.rename(columns={'bag_id': 'original_bag_id',
                           'index': 'bag_id'
                           }, inplace=True)
all_bags['bag_id'] = all_bags['bag_id'] + 1
all_bags['bag_id'] = all_bags['bag_id'].astype(str)

all_bags
```

```python
all_bags.to_csv('../data/finalized_datasets/bags.csv', index=False)
```
