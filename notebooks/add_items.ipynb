```python
import pandas as pd 
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
all_products = pd.read_excel(
    '/project/data/CFTP Test Item Inventory with Dimensions - All Trials.xlsx',
    sheet_name=None,
    skiprows=3
)
products = all_products['Item Inventory - Classes and We']
```

```python
products.columns = [col.lower() for col in products.columns]
products.columns = products.columns.str.replace('.', '').str.replace('@', '').str.replace(',', '')
products.columns = products.columns.str.replace(' ', '_')
products.rename(columns={'item_description_from_trial': 'item_description',
                           'item_id1': 'item_id_10_trials',
                           'certification__time_of_testing': 'certification',
                           'kit_': 'kit',
                           'average_initial_weight_g': 'item_weight',
                           'item_dimensions_compiled': 'item_dimensions',
                           'item_capacity_ml': 'item_capacity'
                           }, inplace=True)
drops = ['item_format', 'material_description',	'material_composition', 'notes', 
            'initial_weight_1',	'initial_weight_2',	'initial_weight_3', 'item_description_refined']
products = products.drop(columns=drops)
products['item_dimensions'] = products['item_dimensions'].str.replace('"', '')
products['weight_unit'] = 'grams'
products['dimensions_unit'] = 'inches'
products['capacity_unit'] = 'milliliters'
new_column_order = ['item_id', 'item_id_10_trials', 'item_name', 'item_description',  'item_sku', 'brand', 
                        'manufacturer', 'material_class_i',	'material_class_ii', 'material_class_iii',	'certification', 
                        'kit', 'item_weight', 'weight_unit', 'item_dimensions', 'dimensions_unit',	
                        'item_capacity', 'capacity_unit']
products = products[new_column_order]
products.to_csv('/project/data/finalized datasets/items.csv', index=False)
```
